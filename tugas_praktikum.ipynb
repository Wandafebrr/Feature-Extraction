# Tugas praktikum
Pada tugas pratikum ini Anda akan menggunakan data "Wisconsin Breast Cancer". Data tersebut terdiri dari 569 data yang digunakan untuk mendiagnonis jenis kanker Malignant (M) dan Benign (B). Tugas Anda adalah,

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
data = 'wbc.csv'
df = pd.read_csv(data)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df[[  'diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
    'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean',
    'symmetry_mean', 'fractal_dimension_mean', 'area_worst', 'perimeter_worst',
    'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst']]
df.head()
```

2. Lakukan proses encoding pada kolom "diagnosis".

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis']) # proses encoding
df.head(10)
```

3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()
df['radius_mean'] = std.fit_transform(df[['radius_mean']])
df['texture_mean'] = std.fit_transform(df[['texture_mean']])
df['perimeter_mean'] = std.fit_transform(df[['perimeter_mean']])
df['area_mean'] = std.fit_transform(df[['area_mean']])
df['smoothness_mean'] = std.fit_transform(df[['smoothness_mean']])
df['compactness_mean'] = std.fit_transform(df[['compactness_mean']])
df['concavity_mean'] = std.fit_transform(df[['concavity_mean']])
df['concave points_mean'] = std.fit_transform(df[['concave points_mean']])
df['symmetry_mean'] = std.fit_transform(df[['symmetry_mean']])
df['fractal_dimension_mean'] = std.fit_transform(df[['fractal_dimension_mean']])
df['area_worst'] = std.fit_transform(df[['area_worst']])
df['perimeter_worst'] = std.fit_transform(df[['perimeter_worst']])
df['compactness_worst'] = std.fit_transform(df[['compactness_worst']])
df['concavity_worst'] = std.fit_transform(df[['concavity_worst']])
df['concave points_worst'] = std.fit_transform(df[['concave points_worst']])
df['symmetry_worst'] = std.fit_transform(df[['symmetry_worst']])
```

```python
df.head()
```

```python
print(f'Jumlah data: {df.shape[0]}')
```

4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python


# Pisahkan fitur (X) dan target (y)
X = df.drop(['diagnosis'], axis=1)  # Hapus kolom yang tidak digunakan
y = df['diagnosis']

# Lakukan stratified split data dengan rasio 80:20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
# Cek ukuran data latih dan data uji
print(f'Jumlah data latih: {X_train.shape[0]}')
print(f'Jumlah data uji: {X_test.shape[0]}')
```
